```python
import pandas as pd
import os

dataframe =  pd.read_csv('../output/Article_IT과학_202012_202012.csv', header=0, encoding='CP949', nrows=2500)
news_title = dataframe.iloc[:, 3].tolist()

if os.path.isfile('../study/dataset/lab11_train_SectionDataset.csv'):
    df = pd.read_csv('../study/dataset/lab11_train_SectionDataset.csv' , names=['title','section'], header=0)

else:
    bodyStruc = {'title': [], 'section': []}
    df = pd.DataFrame( bodyStruc, columns=['title', 'section'] )


print(df.columns)
print(news_title)

df.head()

for item in news_title:
    texts = item
    texts = texts.replace(',', "")
    texts = texts.replace('\'', "")
    texts = texts.replace('\"', "")
    texts = texts.replace('…', " ")
    print(texts)
    oneData = {'title': texts, 'section': '0'}
    df = df.append( oneData, ignore_index=True )

df.to_csv('../study/dataset/lab11_train_SectionDataset.csv', encoding='UTF-8', index=False, header=True)
```
