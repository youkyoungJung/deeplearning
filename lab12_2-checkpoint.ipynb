```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv('../study/dataset/ClusterInput.csv', names=['x', 'y'])
print( df.describe() )
#df2 = df[ ['Sepal.Length', 'Petal.Length'] ] #두개의 칼럼만 사용

model = KMeans( n_clusters=8, random_state=0 )
model.fit(df)

result = model.predict(df)
print(result)
print("개수: ", len(result) )
df.head()
```

```python
plt.scatter( x=df['x'], y=df['y'], c=result )
```

```python
centers = model.cluster_centers_
print(centers)
plt.scatter( x=centers[:,0], y=centers[:,1], marker='D', c='r')
```

```python
plt.scatter( x=df['x'], y=df['y'], c=result )
plt.scatter( x=centers[:,0], y=centers[:,1], marker='D', c='r')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv('../study/dataset/ClusterInput.csv', names=['x', 'y'])
#print( df.describe() )
#df2 = df[ ['Sepal.Length', 'Petal.Length'] ] #두개의 칼럼만 사용

model = KMeans( n_clusters=8, random_state=0 )
model.fit(df)

result = model.predict(df)
centers = model.cluster_centers_

plt.scatter( x=df['x'], y=df['y'], c=result )
plt.scatter( x=centers[:,0], y=centers[:,1], marker='D', c='r')
```
