```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from keras.callbacks import ModelCheckpoint

from keras.models import load_model

import numpy as np
import pandas as pd
import tensorflow as tf
import os


# seed  
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('../Downloads/080228-master/080228-master/deeplearning/dataset/sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

# 10  
n_fold = 5
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

#  accuracy 
accuracy = []
index=0
#  , , 
for train, test in skf.split(X, Y):
   
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])

    MODEL_DIR = './model/'
    if not os.path.exists(MODEL_DIR):
        os.mkdir(MODEL_DIR)

    
    modelpath = "./model/lab7-model-"+str(index)+".hdf5"
   
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
    model.save(modelpath)
   
        
    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)
    index=index+1

#  
print("\n %.f fold accuracy:" % n_fold, accuracy)

```

```python
list_a = list(map(float, accuracy))
print(list_a)
print(np.max(list_a))
maxIndex = list_a.index(np.max(list_a))
print("Max index: ",maxIndex)
```
